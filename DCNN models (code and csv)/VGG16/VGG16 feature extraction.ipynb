```python
from keras.applications.vgg16 import VGG16
from tensorflow.keras.layers import GlobalAveragePooling2D , Reshape
from tensorflow.keras.models import Model
```

```python
base_model = VGG16(weights='imagenet', include_top=True)
```

```python
base_model.summary()
```

```python
x = base_model.layers[-2].output
```

```python
x = Reshape((1,1 , 4096))(x)
```

```python
x = GlobalAveragePooling2D()(x)
```

```python
model = Model(inputs=base_model.input, outputs=x)
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image
import os
import numpy as np
import pandas as pd
```

```python
def extraction(filename , label):
    img = image.load_img(filename, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    features = model.predict(x)
    features = features.tolist()
    features = features[0]
    features.append(label)
    return features
```

```python
categories = os.listdir(f"C:\\Users\\Varad/project work data\\UCMerced_LandUse jpeg format")
array = []

for folder in categories:
    print("label:",folder)
    for file in os.listdir(f"C:\\Users\\Varad/project work data\\UCMerced_LandUse jpeg format\\{folder}"):
        filepath = f"C:\\Users\\Varad/project work data\\UCMerceld_LandUse jpeg format\\{folder}\\{file}"            
        array.append(extraction(filepath,folder))


df = pd.DataFrame(array)
df.to_excel(f'C://Users//Varad//project work data//Result sem 2//vgg16 results.xlsx', index=False)
```
