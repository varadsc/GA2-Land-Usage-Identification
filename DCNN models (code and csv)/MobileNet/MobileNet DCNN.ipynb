```python
from keras.applications import  MobileNet
from tensorflow.keras.layers import GlobalAveragePooling2D , Reshape
from tensorflow.keras.models import Model
```

```python
base_model = MobileNet(weights='imagenet' , include_top = True)
```

```python
base_model.summary()
```

```python
x = base_model.layers[-7].output
```

```python
x = Reshape((7,7,1024))(x)
```

```python
x = GlobalAveragePooling2D()(x)
```

```python
model = Model(inputs=base_model.input , outputs=x)
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image
import os
import numpy as np
import pandas as pd
```

```python
def extraction(filename , label):
    img = image.load_img(filename, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    features = model.predict(x)
    features = features.tolist()
    features = features[0]
    features.append(label)
    return features
```

```python

categories = os.listdir("/kaggle/input/ucmersedfinal/UCMerced_LandUse jpeg format")
array = []

for folder in categories:
    print("label:",folder)
    for file in os.listdir(f"/kaggle/input/ucmersedfinal/UCMerced_LandUse jpeg format/{folder}"):
        filepath = f"/kaggle/input/ucmersedfinal/UCMerced_LandUse jpeg format/{folder}/{file}"            
        array.append(extraction(filepath,folder))


df = pd.DataFrame(array)
df.to_excel(f'/kaggle/working/MobileNet1000 .xlsx', index=False)
```
