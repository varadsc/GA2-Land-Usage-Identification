```python
import pandas as pd
from PIL import Image
from numpy import asarray

def rgb_extracter(filename,label):
    img = Image.open(filename)
    numpydata = asarray(img)
    size_tuple = img.size
#     print(size_tuple)
    resolution = size_tuple[0]*size_tuple[1]
#     print(resolution)
    LR = []
    
        #RED
    r= numpydata.reshape(resolution,3)[:,0]
    r.sort()
    ru = numpydata.reshape(resolution,3)[:int(resolution/2),0]
    rd = numpydata.reshape(resolution,3)[int(resolution/2):,0]
    
    df1 = pd.DataFrame(ru)
    df2 = pd.DataFrame(rd)
    LR=df1.mean()
    UR=df2.mean()

        #GREEN
    g= numpydata.reshape(resolution,3)[:,1]
    g.sort()
    gu = numpydata.reshape(resolution,3)[:int(resolution/2),1]
    gd = numpydata.reshape(resolution,3)[int(resolution/2):,1]
    df1 = pd.DataFrame(gu)
    df2 = pd.DataFrame(gd)
    LG=df1.mean()
    UG=df2.mean()

        #BLUE
    b= numpydata.reshape(resolution,3)[:,2]
    b.sort()
    bu = numpydata.reshape(resolution,3)[:int(resolution/2),2]
    bd = numpydata.reshape(resolution,3)[int(resolution/2):,2]
    df1 = pd.DataFrame(bu)
    df2 = pd.DataFrame(bd)
    LB=df1.mean()
    UB=df2.mean()
    
    return [LR[0],UR[0],LG[0], UG[0],LB[0], UB[0],label]

        
```

```python

def partition(filename,label,n):
    
    img = Image.open(filename)
    numpydata = asarray(img)
    size_tuple = img.size
    resolution = size_tuple[0]*size_tuple[1]
    partitionArray = []
    #for RGB each color
    for i in range(3):
        r= numpydata.reshape(resolution,3)[:,i]
        
        r.sort()
        
#         if i == 0:
#             print("RED")
#         elif i ==1:
#             print("GREEN")
#         else:
#             print("BLUE")
        for j in range(n):
            ru = numpydata.reshape(resolution,3)[int(j*(resolution/n)):int((j+1)*(resolution/n)),i]
            df1 = pd.DataFrame(ru)
            mean=df1.mean()
            partitionArray.append(mean[0])
    
    partitionArray.append(label)
    return partitionArray
```
